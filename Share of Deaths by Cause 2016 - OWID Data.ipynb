```python
"""GLOBAL CAUSE OF DEATH
1. Dalam data ini, aku ingin menemukan 10 negara dengan rate kematian tertinggi untuk masing-masing
   penyebab kematian pada tahun 2016, dinyatakan dalam persen dari total kematian

2. Gunanya adalah untuk melihat, misal mana aja 10 negara dengan rate kematian kanker tertinggi.

3. Misal rate kematian kanker Indonesia 20%, itu artinya 20% orang mati karena kanker. Atau, dari 100 orang yang mati, 20 orang
   mati karena kanker.
"""


import matplotlib.pyplot as plt
import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)





#select.isnull().sum() #ngecek jumlah data kosong (NaN) di tiap kolom
def prepare_data(jenis_kematian):
    data = pd.read_csv("share-of-deaths-by-cause-2016.csv")
    select = data.loc[data.Year==2016] #select data tahun 2016
    select = select[pd.notnull(select["Code"])] #hanya milih data yang kolom "Code" nya nggak kosong. 
    death = select[["Entity", jenis_kematian + " (%)"]]
    death = death.sort_values(jenis_kematian + " (%)") #urutkan data dari rendah ke tinggi
    #death = death.drop(6382) #hilangkan daerah regional e.g: Central Asia, Eastern Europe or WORLD. . 
    death = death.tail(n=10) #ambil data di 10 posisi terakhir. 
    death = death.round(2) #bulatkan ke 2 angka, otomatis akan dipilih data yang bisa dibulatkan

    #langsung bikin grafik
    death.plot.barh(x="Entity", y=jenis_kematian + " (%)", fontsize=16, legend=None, color="rosybrown") #buat grafik batang
    plt.title("10 NEGARA DENGAN PERSENTASE KEMATIAN TERTINGGI KARENA " + jenis_kematian.upper() + " TAHUN 2016", fontsize=20, 
              fontname="monospace")
    plt.xlabel("PERSEN KEMATIAN (% dari total kematian)", fontsize=18, fontname="monospace", labelpad=15)
    plt.ylabel("Negara".upper(), fontsize=18, fontname="monospace", labelpad=15)
    #plt.savefig("fad.png", bbox_inches="tight")
    plt.show()
    print(death)
    
#below dipake untuk dapatin nama2 column(jenis kematian yang ada), kalo udah, capture aja lalu bekukan kodenya
#data = pd.read_csv("share-of-deaths-by-cause-2016.csv") #baca data sek
#for col in data.columns[3:]:
    #print(col)


prepare_data("Diabetes mellitus")

```
